```python
class neuron:
    def __init__(self, input_size, activation):
        self.input_size = input_size
        self.act = activation
        self.w = np.random.uniform(-1,1, len(self.input_size))
    
    def get(self):
        print(f"Input Size: {self.input_size}")
        print(f"Activation Fucntion: {self.act}")
        print(f"Weights: {self.w}")

    def calculate(self, x):
        out = 0
        for i in range(len(self.x)):
            out += self.w[i] * self.x[i]
        return self.act(out)
```

```python
class layer:
    def __init__(self, num_neurons, prev_layer, Input_shape=None):
        self.num_neurons = num_neurons
        self.neurons = []
        self.Input_shape = Input_shape
        self.prev_layer = prev_layer
    
    def generate_neurons(self):
        if self.Input_shape != None:
            for i in range(self.num_neurons):
                self.neurons.append(neuron(self.Input_shape))
        else:
            for i in range(self.num_neurons):
                self.neurons.append(neuron(len(self.prev_layer)))

    def get_neurons(self):
        return self.neurons
```

```python
class activation:
    def __init__(self):
        pass
    
    def tanh(self, x):
        return np.tanh(x)

    def sigmoid(self, x):
        return 1/(1+ np.exp(-x))

    def relu(self, x):
        if x > 0:
            return x
        else:
            return 0

    def linear(self, x):
        return x

class RNN_Module:
    def __init__(self, act1, act2):
        self.w_aa = np.random.uniform(-1,1)
        self.w_ax = np.random.uniform(-1,1)
        self.w_ya = np.random.uniform(-1,1)
        self.b_a = np.random.uniform(-1,1)
        self.b_y = np.random.uniform(-1,1)
        self.g1 = act1
        self.g2 = act2

    def get_input(self, a,x):
        self.a_t_1 = a
        self.x_t = x

    def output(self):
        self.a_t = self.g1(self.a_t_1 * self.w_aa + self.x_t * self.w_ax + self.b_a)
        self.y_t = self.g2(self.a_t * self.w_ya + self.b_y)

        return self.a_t , self.y_t
```

```python
class layer:
    def __init__(self, module_count, act1, act2):
        self.m_c = module_count
        self.act1 = act1
        self.act2 = act2
        self.modules = []
        self.create_module()

    def create_module(self):
        rnn_m = RNN_Module(act1=self.act1, act2=self.act2)
        for i in range(self.m_c):
            self.modules.append(rnn_m)
    
    def forward(self, x, a_0):
        self.a = [a_0]
        self.y = []
        for i in range(self.m_c):
            self.modules[i].get_input(self.a[i],x[i])
            aa, yy = self.modules[i].output()
            self.a.append(aa)
            self.y.append(yy)

        return self.y
```

```python
class model:
    def __init__(self, l_count):
        self.l_c = l_count
        self.layers=[]
        # self.m_c = m_count

    def create_layer(self, m_c, act):
        for i in range(self.l_c):
            self.layers.append(layer(m_c[i], act1=act[i], act2=act[i]))
    
    def forward(self, x, a_0):
        for i in range(self.l_c):
            x = self.layers[i].forward(x, a_0)

        return x
```

```python
class model:
    def __init__(self):
        self.layers=[]

    def add_layer(self, m_c, act1, act2):
        self.layers.append(layer(m_c, act1=act1, act2=act2))
    
    def forward(self, x, a_0):
        for i in range(len(self.layers)):
            x = self.layers[i].forward(x, a_0)

        return x
```
