```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
data = pd.read_csv('android-games.csv')
```

```python
data
```

```python
data.category.unique().shape
```

```python
data.isna().sum()
```

```python
text = ' '.join([cat.split()[1] for cat in data.category])
```

```python
text 
```

```python
wc = WordCloud(collocations=False, background_color='white').generate(text)
```

```python
plt.imshow(wc, interpolation='bilinear');
plt.axis('off');
```

```python
from skimage.io import imread
from wordcloud import ImageColorGenerator
import numpy as np

from skimage.color import rgb2gray
```

```python
img = imread('output.png')
```

```python
mask = np.uint8(rgb2gray(img)>0.5)
```

```python
plt.imshow(mask)
```

```python
wc3 = WordCloud(background_color='white', mask=mask, width=2048, height=1024).generate(text2)

img_color = ImageColorGenerator(img)
wc3.recolor(color_func=img_color)
```

```python
plt.imshow(wc3, interpolation='bilinear');
plt.axis('off');
```

```python
politics_path = '_txt_data/bbc-fulltext/bbc/politics/'

politics_file_list = os.listdir(politics_path)

politics_news = ''
for p in politics_file_list:
    f = open(os.path.join(politics_path, p))
    politics_news = ' '.join(f.read().split()) + ' ' + politics_news

# politics_news = politics_news.replace('Mr', '')
# politics_news = politics_news.replace('will', '')
# politics_news = politics_news.replace('say', '')
# politics_news = politics_news.replace('said', '')
politics_wc = WordCloud(collocations=False,
                        stopwords=StopWords,
                        background_color='#FBFCFC').generate(politics_news.lower())

plt.imshow(politics_wc, interpolation='bilinear');
plt.axis('off');
```

```python
business_path = '_txt_data/bbc-fulltext/bbc/business/'

business_file_list = os.listdir(business_path)

business_news = ''
for p in business_file_list:
    f = open(os.path.join(business_path, p))
    business_news = ' '.join(f.read().split()) + ' ' + business_news
    
# business_news = business_news.replace('Mr', '')
# business_news = business_news.replace('will', '')
# business_news = business_news.replace('say', '')
# business_news = business_news.replace('said', '')
# business_news = business_news.replace('year', '')

business_wc = WordCloud(collocations=False,
                        stopwords=StopWords,
                        background_color='#FBFCFC').generate(business_news.lower())

plt.imshow(business_wc, interpolation='bilinear');
plt.axis('off');
```

```python
from d3graph import d3graph, vec2adjmat
```

```python
s = ['node 1', 'node 1', 'node 1', 'node 2']
t = ['node 2', 'node 3', 'node 4', 'node 4']
w = [2, 5, 10, 20]

adjmat = vec2adjmat(source=s, target=t, weight=w)

d3 = d3graph()
d3.graph(adjmat)
d3.set_node_properties(size=[10, 20, 30, 40], color=['#85C1E9', '#F8C471', '#239B56', '#884EA0'])

d3.show()
```
