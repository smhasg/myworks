```python
import numpy as np 
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt 
```

```python
X ,y= load_digits(return_X_y = True )
```

```python
X.shape ,y.shape
```

```python
def show_digit(X,index):
    plt.imshow(np.reshape(X[index,:],(8,8)))
    plt.show()
```

```python
index = 112
print(y[index])
show_digit(X,index)
```

```python
data_dic={}
for i in range (10): 
    data_dic[str(i)] = X[y==i,:]
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(data_dic['0'])
plt.show()
```
