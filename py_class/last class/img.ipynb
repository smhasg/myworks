```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = io.imread('_img/cameraman.jpeg')
```

```python
img1.shape
```

```python
plt.imshow(img1, plt.cm.gray);
```

```python
img2 = io.imread('_img/flower.jpeg')
```

```python
plt.imshow(img2);
```

```python
img2.shape
```

```python
from skimage.color import rgb2gray
```

```python
img3 = rgb2gray(img2)
```

```python
img3.shape
```

```python
plt.imshow(img3, plt.cm.gray);
```

```python
from skimage import filters
```

```python
plt.imshow(filters.sobel(img3), plt.cm.gray);
```

```python
plt.imshow(filters.sobel_v(img3), plt.cm.gray);
```

```python
plt.imshow(filters.sobel_h(img3), plt.cm.gray);
```
