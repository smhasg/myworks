```python
import numpy as np 
import seaborn as sns 

from sklearn.datasets import make_moons
from sklearn.datasets import make_circles

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec  
```

```python
X,y= make_moons(n_samples=200,random_state=1234 , noise=0.01)
```

```python
X.shape
```

```python
plt.scatter(X[y==0,0],X[y==0,1],c= "red")
plt.scatter(X[y==1,0],X[y==1,1],c= "blue")
```

```python
from sklearn.decomposition import PCA, KernelPCA
```

```python
pca_1d = PCA(n_components=1)
X_1d_pca =pca_1d.fit_transform(X)
```

```python
plt.scatter(X_1d_pca[y==0,0],np.zeros(X_1d_pca[y==0,0].shape)+0.01,c="red",alpha=.1)
plt.scatter(X_1d_pca[y==1,0],np.zeros(X_1d_pca[y==1,0].shape)-0.01,c="blue",alpha=.1)
plt.ylim(-0.2,0.2)
```

```python
kpca_1d = KernelPCA(n_components=1,kernel="rbf",gamma = 1)
X_1d_kpca = kpca_1d.fit_transform(X) 
plt.scatter(X_1d_kpca[y==0,0],np.zeros(X_1d_kpca[y==0,0].shape)+0.01,c="red",alpha=.1)
plt.scatter(X_1d_kpca[y==1,0],np.zeros(X_1d_kpca[y==1,0].shape)-0.01,c="blue",alpha=.1)
plt.ylim(-0.2,0.2)
```

```python
from sklearn.datasets import make_swiss_roll 
```

```python
X , y = make_swiss_roll(n_samples=2000,random_state=1234,noise=.02)
```

```python
#X.shape 
# changing in X effect on mds and isomap in botton 
X[:,2]=0.5*X[:,2]
```

```python
fig = plt.figure(figsize=(8,8)) 
ax = fig.add_subplot(projection = '3d')
ax.scatter3D(X[:,0],X[:,1],X[:,2],c=y)
```

```python
from sklearn.manifold import MDS ,Isomap ,TSNE 
```

```python
X_2d_pca = PCA(n_components=2).fit_transform(X)
X_2d_mds = MDS(n_components=2).fit_transform(X)
X_2d_iso = Isomap(n_neighbors= 5, n_components=2).fit_transform(X)
```

```python
plt.scatter (X_2d_pca[:,0],X_2d_pca[:,1], c= y )
```

```python
plt.scatter (X_2d_mds[:,0],X_2d_mds[:,1], c= y )
```

```python
plt.scatter (X_2d_iso[:,0],X_2d_iso[:,1], c= y )
```

# with runnig time and new fig style 


```python
import timeit
```

```python
runnig_time = []

start= timeit.default_timer()
X_2d_pca = PCA(n_components=2).fit_transform(X)
stop = timeit.default_timer()
runnig_time.append(stop-start)

start= timeit.default_timer()
X_2d_mds = MDS(n_components=2).fit_transform(X)
stop = timeit.default_timer()
runnig_time.append(stop-start)

start= timeit.default_timer()
X_2d_tsne = TSNE(n_components=2).fit_transform(X)
stop = timeit.default_timer()
runnig_time.append(stop-start)

start= timeit.default_timer()
X_2d_iso = Isomap(n_neighbors= 5, n_components=2).fit_transform(X)
stop = timeit.default_timer()
runnig_time.append(stop-start)
runnig_time
```

```python
fig = plt.figure(figsize=(20,10))
spec = fig.add_gridspec(ncols=2,nrows=3)

ax0=fig.add_subplot(spec[0,0])
plt.title("PCA")
plt.scatter(X_2d_pca[:,0],X_2d_pca[:,1],c= y ) 

ax1=fig.add_subplot(spec[0,1])
plt.scatter(X_2d_mds[:,0],X_2d_mds[:,1],c= y ) 
plt.title("MDS")

ax2=fig.add_subplot(spec[1,0])
plt.scatter(X_2d_iso[:,0],X_2d_iso[:,1],c= y ) 
plt.title("ISOMAP")

ax3=fig.add_subplot(spec[1,1])
plt.scatter(X_2d_tsne[:,0],X_2d_tsne[:,1],c= y ) 
plt.title("TSNE")

ax4=fig.add_subplot(spec[2,0])
plt.bar(x=["PCA","MDS","tsne","Isomap"] , height=runnig_time)
plt.title("runnig time")

```
